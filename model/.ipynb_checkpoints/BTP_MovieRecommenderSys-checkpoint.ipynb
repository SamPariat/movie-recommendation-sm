```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.dropna(inplace = True)
```

```python
import ast
```

```python
def extract(obj):
    extracted = []
    for i in ast.literal_eval(obj):
        extracted.append(i['name'])
    return extracted
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
movies.head()
```

```python
# Extract the top 5 actors from the movie
def extractCast(obj):
    extractedCast = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            extractedCast.append(i['name'])
            counter += 1
        else:
            break
    return extractedCast
```

```python
movies['cast'] = movies['cast'].apply(extractCast)
```

```python
movies.head()
```

```python
def fetchDirector(obj):
    extractedDirector = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            extractedDirector.append(i['name'])
            break
    return extractedDirector
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_movies_df
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_movies_df.head()
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x: x.lower())
```

```python
new_movies_df.head()
```

```python
new_movies_df['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
```

```python
def stem_text(text):
    final_list = []
    for i in text.split():
        final_list.append(porter_stemmer.stem(i))
    return ' '.join(final_list)
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(stem_text)
```

```python
new_movies_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cnt_vec = CountVectorizer(max_features=5125, stop_words='english')
```

```python
word_vecs = cnt_vec.fit_transform(new_movies_df['tags']).toarray()
```

```python
word_vecs[0]
```

```python
stacked_arrays = np.column_stack((word_vecs[0], cnt_vec.get_feature_names_out()))

# Print the stacked arrays side by side
for row in stacked_arrays:
    if row[0] >= 1:
        print(row[0], row[1])
```

```python
cnt_vec.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_similarity = cosine_similarity(word_vecs)
```

```python
cos_similarity
```

```python
def recommend_top5_movies(movie):
    # Fetch movie index
    movie_index = new_movies_df[new_movies_df['title'] == movie].index[0]
    distances = cos_similarity[movie_index]
    # Add indexes to each of the movies in the dataframe,
    # then sort by the cosine similarity values
    # then find the top 5 (excludes the current movie)
    top_movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    returned_movies = []
    for i in top_movies_list:
        returned_movies.append([new_movies_df.iloc[i[0]].movie_id, new_movies_df.iloc[i[0]].title])
    
    return returned_movies
```

```python
recommend_top5_movies('Interstellar')
```
